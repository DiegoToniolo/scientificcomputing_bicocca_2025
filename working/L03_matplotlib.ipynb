# Q7: Mandelbrot set

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica"
})

def mandelbrot(c, max_it = 10):
    z = np.zeros(c.shape, dtype = np.complex128)
    it = np.zeros(c.shape, dtype = int)

    for i in range(max_it):
        z = z**2 + c
        it += np.abs(z) <= 2
    return it
```

```python
N = 1000
maxit = 10

# Picture
x = np.linspace(-2, 2, N)
y = np.linspace(-2, 2, N)

xv, yv = np.meshgrid(x, y, indexing="xy")
c = xv + 1j*yv
iterations = mandelbrot(c, maxit)

# Zoom
x = np.linspace(-2, -1.5, N)
y = np.linspace(-0.25, 0.25, N)

xv_zoom, yv_zoom = np.meshgrid(x, y, indexing="xy")
c_zoom = xv_zoom + 1j*yv_zoom
iterations_zoom = mandelbrot(c_zoom, maxit)
```

```python
fig = plt.figure(figsize = (5, 10))

ax1 = fig.add_subplot(211)
ax1.contourf(xv, yv, iterations, levels = np.arange(maxit+1))
ax1.xaxis.set_label_text(r"$\mathrm{Re}\{c\}$")
ax1.yaxis.set_label_text(r"$\mathrm{Im}\{c\}$")
ax1.set_title(r"${\rm The Mandelbrot set}$", fontsize=20)

ax2 = fig.add_subplot(212)
ax2.contourf(xv_zoom, yv_zoom, iterations_zoom, levels=np.arange(maxit+1))
ax2.xaxis.set_label_text(r"$\mathrm{Re}\{c\}$")
ax2.yaxis.set_label_text(r"$\mathrm{Im}\{c\}$")
ax2.set_title("The Mandelbrot set: zoom", fontsize=20)

fig.tight_layout()
```
